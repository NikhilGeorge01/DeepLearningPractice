```python
import numpy as np

class Perceptron:
    def __init__(self, n, lr=0.1):
        self.w, self.b, self.lr = np.random.randn(n), np.random.randn(), lr

    def predict(self, X):
        return (np.dot(X, self.w) + self.b >= 0).astype(int)

    def fit(self, X, y, epochs=100):
        for _ in range(epochs):
            for xi, yi in zip(X, y):
                err = yi - (np.dot(xi, self.w) + self.b >= 0)
                self.w += self.lr * err * xi
                self.b += self.lr * err
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
aY = np.array([0,0,0,1])
oY = np.array([0,1,1,1])

aMod = Perceptron(n = 2)
oMod = Perceptron(n = 2)
aMod.fit(X,aY)
oMod.fit(X, oY)
print(aMod.predict(X), oMod.predict(X))
```

```python
xrY = np.array([0,1,1,0])
xrMod = Perceptron(n=2)
xrMod.fit(X, xrY, epochs=100)
print(xrMod.predict(X))
```

```python
class MPNeuron:
    def __init__(self, threshold=2):
        self.threshold = threshold

    def predict(self, X):
        return (np.sum(X, axis=1) >= self.threshold).astype(int)
```

```python
class PerceptronNoB:
    def __init__(self, n, lr=0.1):
        self.w, self.lr = np.random.randn(n), lr

    def predict(self, X):
        return (np.dot(X, self.w) >= 0).astype(int)

    def fit(self, X, y, epochs=100):
        for _ in range(epochs):
            for xi, yi in zip(X, y):
                err = yi - (np.dot(xi, self.w) >= 0)
                self.w += self.lr * err * xi
```

```python
X = np.array([
    [1, 1, 1, 0.9],  # Nolan + Damon + Thriller + High rating
    [0, 1, 1, 0.8],
    [1, 0, 0, 0.6],
    [0, 0, 0, 0.4],
    [1, 1, 0, 0.7],
    [0, 1, 0, 0.5]
])
y = np.array([1,1,1,0,1,0])
```

```python
m1 = MPNeuron()
m2 = PerceptronNoB(n=4)
m3 = Perceptron(n=4)

m2.fit(X,y)
m3.fit(X,y)

print(m1.predict(X))
print(m2.predict(X))
print(m3.predict(X))
```

```python
import numpy as np

def step(x):
    return (x >= 0).astype(int)

class BooleanMLP:
    def __init__(self, output_truth):
        # output_truth = [y00, y01, y10, y11]
        self.output_truth = output_truth

    def forward(self, X):
        x1, x2 = X[:,0], X[:,1]

        h1 = step(-x1 - x2 + 1.5)
        h2 = step(-x1 + x2 - 0.5)
        h3 = step(x1 - x2 - 0.5)
        h4 = step(x1 + x2 - 1.5)

        H = np.stack([h1,h2,h3,h4], axis=1)

        # Output weights chosen from truth table
        w = np.array(self.output_truth)

        y = step(np.dot(H, w) - 0.5)
        return y
```
